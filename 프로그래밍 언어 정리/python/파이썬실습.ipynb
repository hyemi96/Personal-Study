# 변수의 필요성(원뿔 부피 계산산)

```python
# 원뿔의 반지름과 높이
반지름 = 40
높이 = 30

print(1/3 * 3.14 * 반지름 ** 2 * 높이) 
# 이러식으로 변수를 사용하지 않으면 잘못 계산할 수 있다.
# ** 의 의미는 거듭제곱

# 원뿔의 부피 계산
부피 = 1/3 * 3.14 * 반지름 ** 2 * 높이

print(부피)
```

# 원뿔 부피 계산 프로그램 + 선택 구조

```python
# 원뿔의 반지름과 높이
반지름 = int(input("반지름을 입력해주세요 : "))
높이 = int(input("높이를 입력해주세요 : "))

if 반지름 > 0 and 높이 > 0 :
  # 원뿔의 부피 계산
  부피 = 1/3 * 3.14 * 반지름 ** 2 * 높이
  print("원뿔의 부피는",round(부피), "입니다.")
else :
    print("잘못된 값이 입력되었습니다.")
```
